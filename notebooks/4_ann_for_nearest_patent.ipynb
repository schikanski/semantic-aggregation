```python
from annoy import AnnoyIndex
import pickle
```

```python
seed_list = ['hair_dryer', 'video_codec', 'diesel', "contact_lens", "contact_lens_us_c", "3d_printer"]
src_dir = '.'
seeds = [seed_list[1]]
seed_name = seeds[0]

with open(src_dir + "/data/" + seed_name + "/pre_embedding.pkl", 'rb') as infile:
    df = pickle.load(infile)
```

```python
f = df['document_embeddings'].iloc[0].shape[0]
t = AnnoyIndex(f, 'euclidean')
for idx, val in enumerate(df['document_embeddings']):
    t.add_item(idx, val.squeeze())

t.build(10)
```

```python
patent_idx = 50

result = t.get_nns_by_item(patent_idx, 200)

ranked_df = df.iloc[result]
```
